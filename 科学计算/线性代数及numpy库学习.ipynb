## 特征值特征向量
### 数学描述
#### 定义
$$
\mathrm{设A是n阶方阵，若存在n维非零向量x，使}
$$
$$
\mathrm{Ax}=\mathrm{\lambda x}
$$
$$
\mathrm{则称数\lambda 为A的特征值，x为A属于\lambda 的特征向量}
$$
#### 特征空间
$$
\mathrm{将方程}
$$
$$
\mathrm{Ax}=\mathrm{\lambda x}
$$
$$
\mathrm{改写为}
$$
$$
\left( \mathrm{A}-\mathrm{\lambda I} \right) \mathrm{x}=0
$$
$$
\mathrm{其解空间N}\left( \mathrm{A}-\mathrm{\lambda I} \right) \mathrm{称为A属于\lambda 的特征空间}
$$
#### 特征方程
$$
\mathrm{定义}
$$
$$
\det \left( \mathrm{A}-\mathrm{\lambda I} \right) =0
$$
$$
\mathrm{为A的特征方程，特征方程的根称为特征根，也即A的特征值}
$$

# 向量空间
## 线性无关
$$
\mathrm{对\,\,}\mathbb{R}^{\mathrm{n}}\mathrm{中的k个向量的集合}
$$
$$
\left\{ \mathrm{v}_1,\mathrm{v}_2,\cdots ,\mathrm{v}_{\mathrm{k}} \right\} 
$$
$$
\mathrm{以及k个实数}
$$
$$
\mathrm{\lambda }_1,\mathrm{\lambda }_2,\cdots ,\mathrm{\lambda }_{\mathrm{k}}
$$
$$
\mathrm{若仅当}
$$
$$
\mathrm{\lambda }_1=\mathrm{\lambda }_2=\cdots =\mathrm{\lambda }_{\mathrm{k}}=0
$$
$$
\mathrm{时才有}
$$
$$
\mathrm{\lambda }_1\mathrm{v}_1+\mathrm{\lambda }_2\mathrm{v}_2+\cdots +\mathrm{\lambda }_{\mathrm{k}}\mathrm{v}_{\mathrm{k}}=0
$$
$$
\mathrm{称这k个向量是线性无关的}
$$

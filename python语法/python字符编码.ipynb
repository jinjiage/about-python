    1、unicode是所有字符的全集，包括ascii，中文及各种语系
    2、utf-8、gb2312、gbk等是各种字节编码，为什么有字节编码？
       > 节省内存和带宽，常用字符用最短的值来表示,例如utf-8是变长编码，英文用1字节，中文用3字节；
       > 不同语系，按使用频度来编码，常用的短些，不常用的长些；
       > 不同编码有重叠的部分，这部分就是造成乱码的原因；
